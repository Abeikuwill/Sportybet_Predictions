```python
import numpy as np

#The problem  I am having with the script is that the coeffiecients change during computation, you can compare the printouts and the real values
#Some are automatically rounded, and some are truncated which i dont want. I waant all values to stay intact

# This is a script to compute the values of the coefficients with the user input to generate outcomes
# The outcomes will be filtered to remove the ones outside the range for use (evaluation)
# That same outcome will be sorted in descending order.
# The decision of the outcome will be based on both the filtered result and the sorted results arrangement.

# I will make edits in some parts before integration.

# Define a class for the input data
class BettingData:
    def __init__(self, Home, Draw, Away, Over_2_5, Under_2_5):
        # Convert to float64 and round to 2 decimal places
        self.home_odds = round(float(np.float64(Home)), 2)
        self.draw_odds = round(float(np.float64(Draw)), 2)
        self.away_odds = round(float(np.float64(Away)), 2)
        self.over_2_5_odds = round(float(np.float64(Over_2_5)), 2)
        self.under_2_5_odds = round(float(np.float64(Under_2_5)), 2)
        
# Define coefficients for each type of outcome, rounded to 10 decimal places
coefficients_home = np.array([
    round(-0.0249296730, 10),
    round(0.1157786990, 10),
    round(-0.0367651000, 10),
    round(0.2926870140, 10),
    round(0.1094136140, 10)
], dtype=np.float64)

coefficients_draw = np.array([
    round(-0.0916892420, 10),
    round(0.3345149460, 10),
    round(-0.0824050970, 10),
    round(0.2492194130, 10),
    round(-0.0891180500, 10)
], dtype=np.float64)

coefficients_away = np.array([
    round(-0.0155225780, 10),
    round(0.0894167700, 10),
    round(-0.0233083660, 10),
    round(0.2394321700, 10),
    round(0.1610767770, 10)
], dtype=np.float64)

coefficients_gg = np.array([
    round(-0.0240585870, 10),
    round(0.1333595490, 10),
    round(-0.0378359260, 10),
    round(0.2352813100, 10),
    round(0.1260578330, 10)
], dtype=np.float64)

coefficients_over_2_5 = np.array([
    round(-0.0139963060, 10),
    round(0.1408462300, 10),
    round(-0.0363703600, 10),
    round(0.2308962340, 10),
    round(0.1049659770, 10)
], dtype=np.float64)

# Define functions to calculate different outcomes
def calculate_home(betting_data):
    home_odds_array = np.array([round(betting_data.home_odds, 10)] * len(coefficients_home), dtype=np.float64)
    result = np.dot(coefficients_home, home_odds_array)
    # Display full precision. I will remove the print statements during integration
    print(f"Home Odds Array: {home_odds_array}")
    print(f"Home Coefficients: {coefficients_home}")
    print(f"Home Calculation Result: {round(result, 10):.10f}")   
    return round(result, 10)

def calculate_draw(betting_data):
    draw_odds_array = np.array([round(betting_data.draw_odds, 10)] * len(coefficients_draw), dtype=np.float64)
    result = np.dot(coefficients_draw, draw_odds_array)
    # Display full precision. I will remove the print statements during integration
    print(f"Draw Odds Array: {draw_odds_array}")
    print(f"Draw Coefficients: {coefficients_draw}")
    print(f"Draw Calculation Result: {round(result, 10):.10f}")   
    return round(result, 10)

def calculate_away(betting_data):
    away_odds_array = np.array([round(betting_data.away_odds, 10)] * len(coefficients_away), dtype=np.float64)
    result = np.dot(coefficients_away, away_odds_array)
    # Display full precision. I will remove the print statements during integration
    print(f"Away Odds Array: {away_odds_array}")
    print(f"Away Coefficients: {coefficients_away}")
    print(f"Away Calculation Result: {round(result, 10):.10f}") 
    return round(result, 10)

def calculate_gg(betting_data):
    gg_odds_array = np.array([round(betting_data.under_2_5_odds, 10)] * len(coefficients_gg), dtype=np.float64)
    result = np.dot(coefficients_gg, gg_odds_array)
    # Display full precision. I will remove the print statements
    print(f"GG Odds Array: {gg_odds_array}")
    print(f"GG Coefficients: {coefficients_gg}")
    print(f"GG Calculation Result: {round(result, 10):.10f}")   
    return round(result, 10)

def calculate_over_2_5(betting_data):
    over_2_5_odds_array = np.array([round(betting_data.over_2_5_odds, 10)] * len(coefficients_over_2_5), dtype=np.float64)
    result = np.dot(coefficients_over_2_5, over_2_5_odds_array)
    # Display full precision. I will remove the print statements
    print(f"Over 2.5 Odds Array: {over_2_5_odds_array}")
    print(f"Over 2.5 Coefficients: {coefficients_over_2_5}")
    print(f"Over 2.5 Calculation Result: {round(result, 10):.10f}")  
    return round(result, 10)

# Create an instance of BettingData. Edit this to the real web elements in selenium when integrating
def get_user_input():
    Home = np.float64(input("Enter Home Odds: "))
    Draw = np.float64(input("Enter Draw Odds: "))
    Away = np.float64(input("Enter Away Odds: "))
    Over_2_5 = np.float64(input("Enter Over 2.5 Odds: "))
    Under_2_5 = np.float64(input("Enter Under 2.5 Odds: "))
    return BettingData(Home, Draw, Away, Over_2_5, Under_2_5)

betting_data = get_user_input()

# Calculate different outcomes
home_result = calculate_home(betting_data)
draw_result = calculate_draw(betting_data)
away_result = calculate_away(betting_data)
gg_result = calculate_gg(betting_data)
over_2_5_result = calculate_over_2_5(betting_data)

# Evaluate results
def evaluate_home(home_value):
    range1 = (round(0.943378894, 10), round(0.965300849, 10))
    range2 = (round(1.028596461, 10), round(1.06997017, 10))
    result = "Home" if (range1[0] <= home_value <= range1[1]) or (range2[0] <= home_value <= range2[1]) else "SKIP"
    return result

def evaluate_draw(draw_value):
    range1 = (round(0.943041151, 10), round(0.969241237, 10))
    range2 = (round(1.031015099, 10), round(1.066588876, 10))
    result = "Draw" if (range1[0] <= draw_value <= range1[1]) or (range2[0] <= draw_value <= range2[1]) else "SKIP"
    return result

def evaluate_away(away_value):
    range1 = (round(0.9237984600, 10), round(0.9607534460, 10))
    range2 = (round(1.0416107450, 10), round(1.0855130600, 10))
    result = "Away" if (range1[0] <= away_value <= range1[1]) or (range2[0] <= away_value <= range2[1]) else "SKIP"
    return result

def evaluate_gg(gg_value):
    range1 = (round(0.9597137090, 10), round(0.9741679880, 10))
    range2 = (round(1.0154605770, 10), round(1.0422988860, 10))
    result = "GG" if (range1[0] <= gg_value <= range1[1]) or (range2[0] <= gg_value <= range2[1]) else "SKIP"
    return result

def evaluate_over_2_5(over_2_5_value):
    range1 = (round(0.9380674390, 10), round(0.9683560690, 10))
    range2 = (round(1.0308001910, 10), round(1.0629556840, 10))
    result = "Over 2.5" if (range1[0] <= over_2_5_value <= range1[1]) or (range2[0] <= over_2_5_value <= range2[1]) else "SKIP"
    return result

evaluated_results = {
    "Home": evaluate_home(home_result),
    "Draw": evaluate_draw(draw_result),
    "Away": evaluate_away(away_result),
    "GG": evaluate_gg(gg_result),
    "Over 2.5 Goals": evaluate_over_2_5(over_2_5_result)
}

# Prepare results for sorting
non_filtered_results = {
    "Home": home_result,
    "Draw": draw_result,
    "Away": away_result,
    "GG": gg_result,
    "Over 2.5 Goals": over_2_5_result
}

# Sort the results in descending order
sorted_results = sorted(non_filtered_results.items(), key=lambda x: x[1], reverse=True)

# Print sorted results
print("Sorted Results:")
for name, value in sorted_results:
    print(f"{name}: {round(value, 10):.10f}")  # Display full precision

# Print evaluated results
print("\nEvaluated Results:")
for name, value in evaluated_results.items():
    print(f"{name}: {round(value, 10):.10f}" if isinstance(value, float) else f"{name}: {value}")

# Decision Tree Logic
def decision_tree(evaluated_results, sorted_results):
    # Create a list for the order of evaluated results including "SKIP"
    evaluated_order = [(name, idx + 1) if value != "SKIP" else (name, "SKIP") for idx, (name, value) in enumerate(evaluated_results.items())]
    # Create a list for the order of sorted results
    sorted_order = [(name, idx + 1) for idx, (name, _) in enumerate(sorted_results)]

    # Print the evaluated order and sorted order for debugging
    print("\nEvaluated Order:")
    for name, position in evaluated_order:
        print(f"{name}: {position}")

    print("\nSorted Order:")
    for name, position in sorted_order:
        print(f"{name}: {position}")

    # Define patterns for decision-making. I will add all the patterns later
    pattern1 = ["Home", "Draw", "Away", "GG", "Over 2.5 Goals"]
    pattern2 = ["Draw", "Home", "Away", "GG", "Over 2.5 Goals"]

    # Determine decision based on patterns. I will add the 720 elifs later
    if [name for name, _ in evaluated_order] == pattern1 and [name for name, _ in sorted_order] == pattern1:
        return "Decision: Choose Home"
    elif [name for name, _ in evaluated_order] == pattern2 and [name for name, _ in sorted_order] == pattern2:
        return "Decision: Choose Draw"
    else:
        return "Decision: Unable to determine based on patterns."

# Print the decision based on evaluated results and sorted results
decision = decision_tree(evaluated_results, sorted_results)
print(f"\n{decision}")
```
